```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
base_url = "https://slickdeals.net/deals/?page="
all_popular_deals_info = []
for num in range(1, 6):
    my_url = base_url + str(num)
    single_page_html = requests.get(my_url).content
    single_page_doc = BeautifulSoup(single_page_html,'html.parser')
    all_deals = single_page_doc.find_all("div", class_="dealRow")  
    for deal in all_deals:
        deal_title = deal.find("div", class_="dealTitle").text.strip().split('\n')[1]
        slick_deal_url = "https://slickdeals.net" + deal.find("a", class_="track-popularDealLink bp-p-dealLink bp-c-link")["href"]
        deal_info_html = requests.get(slick_deal_url).content
        deal_info_doc = BeautifulSoup(deal_info_html,'html.parser')   
        deal_link_type_1 = deal_info_doc.find("a", class_="dealDetailsOutclickButton dealDetailsMainBlock__outclickButton")
        deal_line_type_2 = deal_info_doc.find("a", class_="seeDealButton")
        if deal_link_type_1:
            deal_link = deal_link_type_1["href"]
        elif deal_line_type_2:
            deal_link = deal_line_type_2["href"]
        else:
            deal_link = "N/A"
        deal_price = deal.find("div", class_="price").get_text(strip=True)
        posted_time = deal.find("div", class_="dealLinks").text.strip().split('\n')[0]
        rating = deal.find("div", class_="ratingCol stats").get_text(strip=True)
        activity_col = deal.find("div", class_="activityCol stats")
        views = activity_col.find("div", class_="activity").text.strip().split('\n')[0]
        comments = activity_col.find_all("div", class_="activity")[1].text.strip().split('\n')[0]
        all_popular_deals_info.append({"title": deal_title, "link": deal_link, "price": deal_price, "posted_time": posted_time, "rating": rating, "views": views, "comments": comments})
```

```python
print(all_popular_deals_info)
```

```python
import pandas as pd
deals_df = pd.DataFrame(all_popular_deals_info)
deals_df['title'] = deals_df['title'].apply(lambda x: [x])
deals_df['link'] = deals_df['link'].apply(lambda x: f'({x})')
deals_df['title'] = deals_df['title'].astype(str) + deals_df['link']
deals_df.drop('link', axis=1, inplace=True)
deals_df
```

```python
output_file = 'popular_deals.csv'
deals_df.to_csv(output_file, index=False, encoding='utf-8')

print(f"Data saved to {output_file}")
```
